```python
# mani 칼럼 생성

mani = []

for i in range(len(pledge)):
    if pledge.vin.iloc[i]:
        if pledge.vin.iloc[i][1] == 'V':
            mani.append('폭스바겐')
        elif pledge.vin.iloc[i][1] == 'A':
            mani.append('아우디')
        elif pledge.vin.iloc[i][1] == 'B':
          mani.append('BMW')
        elif pledge.vin.iloc[i][1] == 'D':
          mani.append('벤츠')
        elif pledge.vin.iloc[i][:2] == 'KP':
          mani.append('쌍용')
        elif pledge.vin.iloc[i][1] == 'M':
          mani.append('현대')
        elif pledge.vin.iloc[i][1] == 'L':
          mani.append('대우')
        elif pledge.vin.iloc[i][1] == 'N':
          mani.append('기아')
        elif pledge.vin.iloc[i][1] == 'P':
          mani.append('포르쉐')
        else:
          mani.append('기타')
    else:
        mani.append(None)
```

```python
pledge['mani'] = mani
pledge.mani.value_counts(normalize= True)
```
