#아이트럭 DB Table

내차사기 아이디(id): id값  
대분류(category): 차종 유형 큰 단위 분류(아이트럭 '형식' 카테고리)  
소분류(sub_category): 차종 유형 작은 단위 분류(아이트럭 '형식'안에 카테고리들의 분류)  
번호판 종류(license_plate) : [자가용 / 영업용]  
톤수(ton): 적재 톤수
브랜드(brand) : ['현대', '벤츠', '타타대우', '볼보', '기아', '만', '스카니아', '이스즈', '이베코', '기타']  
모델(model) : 차종 모델명  
연식(vehicle_year) : 차량이 제조된 년도  
주행거리(mileage) : 총 주행거리  
가격(price) : 천만원 단위  
매물상태(con_sale) : ['거래중', '거래완료', '숨김', '예약중']  
우대매물여부(prefer_sale) : ['Y', 'N']  
특수차량 소분류 명칭(special_vehicle) :  특수차량을 소분류   
실매물 확인여부(check_product) : ['Y', 'N']    
지역1 명칭(state) : 특별시/도 단위 지역    
지역2 명칭(city) : 시/군/구 단위 지역  
거래완료가격(transac_price) : 거래완료된 가격  
거래완료일자(transac_price) : 거래완료된 날짜    
등록일자(reg_date) : 중고차량 등록 일자    
적재함 길이(len) : 적재함 길이(cm)    
적재함 폭(width) : 적재함 폭(cm)  
적재함 높이(height) : 적재함 높이(cm)  
가변축(add_tire) : ['N', 'B', 'F', ''] = ['없음', '뒤축', '앞축', '']  
보링 여부(engine_repair) : ['Y', 'N']

```python
# mani 새로운 제조사 컬럼 생성
import numpy as np

vin_mani_list = []

for value in change.vin.values:
    if value:
        if value[1] == 'V':
            vin_mani_list.append('폭스바겐')    
        elif value[1] == 'A':
            vin_mani_list.append('아우디')
        elif value[1] == 'B':
            vin_mani_list.append('BMW')
        elif value[1] == 'D':
            vin_mani_list.append('벤츠')
        elif (value[1] == 'P') & (value[0] == 'K'):
            vin_mani_list.append('쌍용')
        elif value[1] == 'M':
            vin_mani_list.append('현대')
        elif value[1] == 'L':
            vin_mani_list.append('대우')
        elif value[1] == 'N':
            vin_mani_list.append('기아')
        elif value[1] == 'P':
            vin_mani_list.append('포르쉐')
        elif value[1] == 'R':
            vin_mani_list.append('쌍용')
        else:
            vin_mani_list.append('기타')
    else:
        vin_mani_list.append(np.NaN)

# print(vin_mani_list[:100])

change['mani'] = vin_mani_list
```
