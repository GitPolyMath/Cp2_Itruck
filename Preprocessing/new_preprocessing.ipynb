#신규등록정보 (2017.04 - 2022.07)

- 제원관리번호(manage_id) : 자동차 제작사 혹은 수입사가 관련 기관에 해당 모델에 관한 자세한 제원 정보를 주게 되면 각 모델 별 제원을 관리하는 제원관리 번호가 발급  
    - 제원 : 자동차 설계나 제조의 표준이 되고 형식승인을 받을 때 요구되는 항목  
- 차종명(vehicle_type) : 특수,화물로 구분  
- 차종유형(vehicle_model) : ['견인', '특수작업형', '구난', '일반형-픽업', '일반형-카고', '특수용도형-특수용도형', '밴', '특수용도형-피견인', '덤프', '특수용도형-탱크로리', '특수용도형']  
    - 견인/피견인 : 견인 가능한 차량 /불가차량  
    - 구난 : 견인차  
- 차종분류명(vehicle_cat) : 대/중/소/경/nan  
- 배기량(displacement) : 엔진 내부적으로 피스톤이 움직일 때 흡입, 배출되는 공기의 양  
    - 높으면 엔진의 크기가 크고 소비되는 연료도 많음 (일반적으로) 
- 승차정원수(passenger) : 앉을수 있는 승객 수
- 최대적재량(max_load) : 화물차 짐칸에 실을 수 있는 화물의 최대 분량  
- 차명(vehecle_name) : 차종 이름  
- 모델연도(model_year) : 모델 만들어진 연도
- 사용연료명(fuel) : 사용하는 연료명
- 국산수입구분명(dom_import) : ['국산', '외산']
- 차대차체형상명(vehicle_body) : 트럭형상  
    - 캡 오버 : 차량의 캡이 맨 앞 끝까지 위치한 형식  
    - 보넷 : 엔진이 위치한 보닛이 앞으로 튀어나온 형태 (컨벤셔널타입이라고도 부름)  
    - 세미보넷 : 엔진 룸의 반 정도가 전면으로 돌출된 차량  
- 차체길이(len) 
- 차체너비(width)
- 차체높이(height)
- 축간거리(wheelbase)
- 전윤간거리(front_wb)
- 후윤간거리(back_wb)
- 변속기종류명(transmission) : ['자동', '수동', '반자동', '변속기 없음', '기타', nan, '무단']  
- 차대번호(vin) : 차대번호는 차체 그 자체만의 고유번호  
- 등록관청명(reg_area) : 등록한 특별시/도 지역  
- 용도구분명(purpose) : 용도구분  
    - 관용 : 정부기관 혹은 기업체에서 업무용으로 사용되는 차량 
- 등록일(sale_reg_date) : 등록일자  
- 등록상세명(reg_detail) : ['일반소유용', '매매상품용'] 
- 최초등록일(first_reg_date) : 최초 등록날짜
- 제작년월일(man_date) : 제작 년/월/일
- 회원구분명(member_classi) : ['법인및사업자', '개인']
- 신청구분명(apply_category) : ['신조차 신규등록', '수입차 신규등록', '부활차이전등록', '부활차단순등록']  
- 취득금액(pur_amount) : 차량 구매한 금액 
- 원동기최고출력(max_output) : 마력  
- 연료소비율(consum_rate) : 연료 1리터로 주행할 수 있는 거리(킬로미터)  
- 차량총중량(weight)
- 최초제원관리번호(first_manage_id)
- 원동기RPM값(rpm) : 분당회전수 ( 엔진의 회전수가 높아질수록 토크가 일정하다면 출력도 높아진다. 그러나 동시에 연료 소모와 소음, 진동도 커진다. )  
- 구동방식코드(dri_system) : M X N 으로 표시하며 M 은 바퀴의 (위치)수, N 은 구동 바퀴의 (위치)수 로 표현  
    - 구동 바퀴 : 엔진에서 발생한 동력이 전달 되는 바퀴  
- 최초취득금액(first_pur_amount)

##제조사컬럼(mani) 생성 from 차대번호(vin)

```python
mani = []

for i in range(len(new)):
    if new.vin.iloc[i]:
        if new.vin.iloc[i][1] == 'V':
            mani.append('폭스바겐')
        elif new.vin.iloc[i][1] == 'A':
            mani.append('아우디')
        elif new.vin.iloc[i][1] == 'B':
          mani.append('BMW')
        elif new.vin.iloc[i][1] == 'D':
          mani.append('벤츠')
        elif new.vin.iloc[i][:2] == 'KP':
          mani.append('쌍용')
        elif new.vin.iloc[i][1] == 'M':
          mani.append('현대')
        elif new.vin.iloc[i][1] == 'L':
          mani.append('대우')
        elif new.vin.iloc[i][1] == 'N':
          mani.append('기아')
        elif new.vin.iloc[i][1] == 'P':
          mani.append('포르쉐')
        else:
          mani.append('기타')
    else:
        mani.append(None)
```

```python
new.vin.iloc[0][:2]
```

```python
new['mani'] = mani
new.mani.value_counts(normalize= True)
```

```python
# sale_reg_date
new['sale_reg_date'] = new['sale_reg_date'].astype({'sale_reg_date':'object'})

for i, value in enumerate(new['sale_reg_date'].values):
    new['sale_reg_date'].values[i] = f"{str(value)[:4]}-{str(value)[4:]}-01"
```

```python
# first_reg_date 
new['first_reg_date'] = new['first_reg_date'].astype({'first_reg_date':'object'})

for i, value in enumerate(new['first_reg_date'].values):
    new['first_reg_date'].values[i] = f"{str(value)[:4]}-{str(value)[4:]}-01"
```

```python
# man_date
new.dropna(subset=['man_date'], axis=0, inplace=True)
new['man_date'] = new['man_date'].astype({'man_date':'int64'})
new['man_date'] = new['man_date'].astype({'man_date':'object'})

for i, value in enumerate(new['man_date'].values):
    new['man_date'].values[i] = f"{str(value)[:4]}-{str(value)[4:]}-01"
```

```python
# 모델연도 datetime형태로 변경
new['sale_reg_date'] = pd.to_datetime(new['sale_reg_date'],format='%Y-%m-%d')
new['first_reg_date'] = pd.to_datetime(new['first_reg_date'],format='%Y-%m-%d')
new['man_date'] = pd.to_datetime(new['man_date'],format='%Y-%m-%d')
```
