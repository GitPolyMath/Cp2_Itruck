### manage_id / engine / vin 을 토대로 결측치 제거 
### 해당 함수 하나로 만들 수 있을 것 같긴한데 헷갈림..

```python
vcmi = transfer[(transfer['vehicle_cat'].isna())]['manage_id'].tolist()
for i in range(len(vcmi)):
    t = transfer.loc[(transfer['manage_id'] == vcmi[i])]['vehicle_cat']
    if len(t.unique()) <= 1:
        continue
    transfer.loc[(transfer['vehicle_cat'].isna()) & (transfer['manage_id'] == vcmi[i]),'vehicle_cat'] = t.mode()[0]
```

```python
vcengine = transfer[(transfer['vehicle_cat'].isna())]['engine'].tolist()
for i in range(len(vcengine)):
    t = transfer.loc[(transfer['engine'] == vcengine[i])]['vehicle_cat']
    if len(t.unique()) <= 1:
        continue
    transfer.loc[(transfer['vehicle_cat'].isna()) & (transfer['engine'] == vcengine[i]),'vehicle_cat'] = t.mode()[0]
```

```python
vcvin = transfer[(transfer['vehicle_cat'].isna())]['vin'].tolist()
for i in range(len(vcvin)):
    t = transfer.loc[(transfer['vin'] == vcvin[i])]['vehicle_cat']
    if len(t.unique()) <= 1:
        continue
    transfer.loc[(transfer['vehicle_cat'].isna()) & (transfer['vin'] == vcvin[i]),'vehicle_cat'] = t.mode()[0]
```

```python
transfer[(transfer['vehicle_cat'].isna()) & (~transfer['vehicle_name'].str.contains('트레일'))]
```

```python
def fillfunc(col,tool, transfer=transfer):
    mi = set(transfer[(transfer[col].isna())][tool].tolist())
    for i in mi:
        t = transfer.loc[(transfer[tool] == i)][col]
        if len(t.unique()) <= 1:
            continue
        transfer.loc[(transfer[col].isna()) & (transfer[tool] == i), col] = t.mode()[0]
```

```python
nacol = ['displacement','passenger','max_load','vehicle_name','model_year','fuel','dom_import','len','width','transmission','consum_rate','weight','rpm']
tool = ['manage_id','engine','vin']
```

```python
set(transfer[(transfer['passenger'].isna())]['manage_id'].tolist())
```

```python
for i in nacol:
    for k in tool:
        fillfunc(i,k)
```

```python
def going_to_light(df):
    for i in df.columns:
        if df[i].dtype == 'int64':
            df[i] = df[i].astype('int32')
        if df[i].dtype == 'float64':
            df[i] = df[i].astype('float32')
        if df[i].dtype == 'object':
            df[i] = df[i].astype('category')
    return df
```
