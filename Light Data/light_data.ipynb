## 메모리 최적화(int64 -> int32, float64 -> float32, object -> category)

```python
def going_to_light(df):
    for i in df.columns:
        if df[i].dtype == 'int64':
            df[i] = df[i].astype('int32')
        if df[i].dtype == 'float64':
            df[i] = df[i].astype('float32')
        if df[i].dtype == 'object':
            df[i] = df[i].astype('category')
    return df
```

```python
registration = going_to_light(registration)
change_1 = going_to_light(change_1)
change_2 = going_to_light(change_2)
cancel_1 = going_to_light(cancel_1)
cancel_2 = going_to_light(cancel_2)
new_1 = going_to_light(new_1)
new_2 = going_to_light(new_2)
transfer_1 = going_to_light(transfer_1)
pledge_1 = going_to_light(pledge_1)
pledge_2 = going_to_light(pledge_2)
```

# Parquet 파일 생성

```python
import numpy as np
def to_nan(x):
    if x == '-':
        x = np.NaN
    if type(x) == str:
        x = float(x.strip())
    return x
```

```python
cancel["원동기최고출력"] = cancel["원동기최고출력"].apply(to_nan)
cancel["원동기RPM값"] = cancel["원동기RPM값"].apply(to_nan)
```

```python
pledge["원동기최고출력"] = pledge["원동기최고출력"].apply(to_nan)
pledge["원동기RPM값"] = pledge["원동기RPM값"].apply(to_nan)
```

```python
import fastparquet

registration.to_parquet('registration.parquet', engine = 'fastparquet', compression = 'gzip')
change.to_parquet('change.parquet', engine = 'fastparquet', compression = 'gzip')
cancel.to_parquet('cancel.parquet', engine = 'fastparquet', compression = 'gzip')
new.to_parquet('new.parquet', engine = 'fastparquet', compression = 'gzip')
transfer_1.to_parquet('transfer.parquet', engine = 'fastparquet', compression = 'gzip')
pledge.to_parquet('pledge.parquet', engine = 'fastparquet', compression = 'gzip')
```
